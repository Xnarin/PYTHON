```python
def add_many(*args): # 매개변수 앞에 *을 붙이면 입력값을 전부 모아서 튜플로 만들어 준다.
    total = 0 # 초기값 설정
    for i in args: # 함수 안에서 사용할 때, *없이 변수명
        total = total + i
    return total
```

```python
add_many(1,2,3)
```

```python
def div(a,b):
    try:
        return a/b
    except ZeroDivisionError:
        return '계산식 표현 불가'
```

```python
def div(a,b):
    try:
        return a/b
    except ZeroDivisionError as Error1:
        return '{}로 인한 계산식 표현 불가' .format(Error1)
div(1,0)
```
